```python
import numpy as np
import matplotlib.pyplot as plt
import random 
```

```python
#Generar una Gaussiana 

def gaussian(x, mu, sigma):
    return np.exp(-np.power(x - mu, 2.) / (2 * np.power(sigma, 2.)))

x = np.arange(-20, 20, 0.1)
x

g = gaussian(x, 0,4) #gaussiana centrada en 0 y con desviancion estandar 4
g
#graficar 
plt.plot(x,g)

```

```python
#aplicando la FFT con numpy y graficando

f = np.fft.fft(g)
f

fft = plt.plot(x,f)

```

```python
#para la tranformada inversa

fi = np.fft.ifft(f)
fi

ffi = plt.plot(x,fi)
```

```python
fig, axs = plt.subplots(3,1, figsize=(15, 10))

axs[0].plot(g)
axs[1].plot(f)
#axs[1].set_xlim(-20,20)
axs[2].plot(fi)



```
